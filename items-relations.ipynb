```python
nodes = []
count = 0
    
for relation in relations:
    print(relation)


    ## Obtener los metadatos de la obra
    obra_id = relation["subject"]["id"]
    obra_url = base_url + "/items/%s"% obra_id
    print(obra_url)

    obra_item = requests.get(obra_url).json()
    #print(obra_item)

    # campos a buscar en los metadatos
    obra_title = ""
    obra_date = ""
    obra_format = ""
    obra_medium = ""

    #iteracion por los textos de elemento, o valores de los metadatos
    for element_text in obra_item["element_texts"]:
        if "Title" == element_text["element"]["name"]:
            obra_title = element_text["text"]

        if "Date" == element_text["element"]["name"]:
            obra_date = element_text["text"]

        if "Format" == element_text["element"]["name"]:
            obra_format = element_text["text"]

        if "Medium" == element_text["element"]["name"]:
            obra_medium = element_text["text"]
    
    
        obra_node = {
            "id": obra_id,
            "url": obra_url,
            "title": obra_title,
            "date" : obra_date,
            "format" : obra_format,
            "medium" : obra_medium,
            "type" : "obra"
        }
        
        
    ## Obtener los metadatos del referente
    referente_id = relation["object"]["id"]
    referente_url = base_url + "/items/%s"% referente_id

    referente_item = requests.get(referente_url).json()


    referente_title = ""
    referente_date = ""
    referente_publisher = ""
    

    #iteracion por los textos de elemento, o valores de los metadatos
    for element_text in referente_item["element_texts"]:
        if "Title" == element_text["element"]["name"]:
            referente_title = element_text["text"]

        if "Date" == element_text["element"]["name"]:
            referente_date = element_text["text"]

        if "Publisher" == element_text["element"]["name"]:
            referente_publisher = element_text["text"]

    # Fila de la tabla

    referente_node = {
               "id" : referente_id,
               "url" : referente_url,
               "title" : referente_title,
               "date" : referente_date,
               "publisher" : referente_publisher
    }
    
    nodes.append(obra_node)
    nodes.append(referente_node)
    
    count = count + 1

    
print(count)

json_object["links"] = links
json_object["nodes"] = nodes

    
with open(json_out, 'w') as jsonfile:
    json.dump(json_object, jsonfile)

```
